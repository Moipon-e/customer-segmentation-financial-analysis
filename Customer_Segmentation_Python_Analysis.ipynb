```
df = pd.read_excel('Customer Segmentation Financial Behaviour.xlsx', sheet_name='Customer Data')

df.head()
```

```
# Revenue by Customer Segment
df.groupby('Customer Segment')['Transaction_Amount'].sum()
```

```
# Average behaviour by segment
df.groupby('Customer Segment')[['Transaction_Amount', 'Transaction_Count']].mean()
```

```
import matplotlib.pyplot as plt

# Revenue by segment
revenue = df.groupby('Customer Segment')['Transaction_Amount'].sum()

# Plot
revenue.plot(kind='bar')

plt.title('Revenue by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Total Revenue')

plt.show()
```

```
# Average behaviour by segment
behaviour = df.groupby('Customer Segment')[['Transaction_Amount', 'Transaction_Count']].mean()

# Plot
behaviour.plot(kind='bar')

plt.title('Customer Behaviour by Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average Values')

plt.show()
```
